```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#loading the raw data
population_raw = pd.read_csv('./data/YearlyBeePopTrend.csv', thousands=',')
population_raw['Year'].unique() #checking what year's are included in the data
```

```python
#converting data to appropriate datay type
population_raw['Year'] = population_raw['Year'].astype(int)
population_raw['Value'] = population_raw['Value'].astype(float)
```

```python
#Selecting annual data
marketing_year_data = population_raw[population_raw['Period'] == 'MARKETING YEAR']
```

```python
marketing_year_data
```

```python
#checking what information is available 
marketing_year_data['Year'].unique()
marketing_year_data['Data Item'].unique()
```

```python
pop = marketing_year_data #reassign variable
```

```python
#creating a plot
fig, ax = plt.subplots(figsize = (9, 9))

#plotting a scatterplot
ax.scatter(x=pop['Year'], y=pop['Value'])

#calculating line of best fit
b, a = np.polyfit(pop['Year'], pop['Value'], deg=1)
beta, alpha = np.polyfit(range(len(pop['Year'])), pop['Value'][::-1], deg=1)
    
#Calculating R-squared value
corr_matrix = np.corrcoef(range(len(pop['Year'])), pop['Value'][::-1])
corr = corr_matrix[0,1]
R_sq = corr**2

ax.plot(pop['Year'], a + b * pop['Year'], color="k", lw=2.5, label=f'{round(alpha, 2)} + {round(beta, 2)}*x, R^2={round(R_sq, 3)}')

#Configuring chart labels
ax.set_title(f"Number of Honey Bee Colonies in the US from 1987 - 2022", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("Years", fontsize=14)
ax.set_ylabel("Number of Colonies", fontsize=14)
ax.legend(loc="upper right", fontsize=14)

```
